```python
#Function
from docplex.mp.model import Model
#Z12 Range Function
def solve_Z12_range():
    """
    Solve PJK submodel to find Z12_min and Z12_max.
    """
    mdl = Model(name="PJK_Submodel")

    # Define decision variables
    x = mdl.binary_var_dict(SR, name='x_')                                                                                        # x[r]
    y = mdl.binary_var_dict(((r, f, m) for r in SR for f in SF for m in SM), name='y_')                                           # y[r,f,m]

    # Define constraints
    for s in SS:
        mdl.add_constraint(mdl.sum(alpha[r][s] * x[r] for r in SR) >= 1, ctname=f"min_coverage_{s}")                              # Constraint 5
    mdl.add_constraint(mdl.sum(x[r] for r in SR) <= eta, ctname="max_routes")                                                     # Constraint 6
    for r in SR:
        for f in SF:
            for m in SM:
                mdl.add_constraint(m * y[r, f, m] <= lamda[r], ctname=f"cap_limit_{r}_{f}_{m}")                                   # Constraint 7
    for e in SE:
        mdl.add_constraint(mdl.sum(o[m] * beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM)                            # Constraint 8
                           >= omega[e], ctname=f"min_load_{e}")
    for e in SE:
        mdl.add_constraint(mdl.sum(beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM) >= ae, ctname=f"min_flow_{e}")    # Constraint 9a
        mdl.add_constraint(mdl.sum(beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM) <= be, ctname=f"max_flow_{e}")    # Constraint 9b
    for r in SR:
        mdl.add_constraint(mdl.sum(y[r, f, m] for f in SF for m in SM) <= F * x[r], ctname=f"route_link_{r}")                     # Constraint 20
    #for f in range(1, F):
        #mdl.add_constraint(
            #mdl.sum(y[r, f + 1, m] for r in SR for m in SM) <= mdl.sum(y[r, f, m] for r in SR for m in SM),
            #ctname=f"sequential_flow_{f}")                                                                                        # Constraint 21
    for f in SF:
        mdl.add_constraint(mdl.sum(y[r, f, m] for r in SR for m in SM) <= 1, ctname=f"max_flow_per_freq_{f}")                     # Constraint 22

    # Define objective
    Z12 = mdl.sum(c2[m] * g[r] * y[r, f, m] for r in SR for f in SF for m in SM)

    # Solve for Z12_min using interior point method
    mdl.parameters.lpmethod.set(method)  # This specifies the interior-point method for CPLEX
    mdl.minimize(Z12)
    solution_min = mdl.solve()
    if not solution_min:
        print("Model is infeasible when minimizing.")
        return None, None
    Z12_min = solution_min.get_objective_value()

    # Solve for Z12_max using interior point method
    mdl_max = mdl.clone()
    mdl_max.parameters.lpmethod.set(method)  # Set interior-point method for the clone model
    mdl_max.maximize(Z12)
    solution_max = mdl_max.solve()
    if not solution_max:
        print("Model is infeasible when maximizing.")
        return None, None
    Z12_max = solution_max.get_objective_value()

    return Z12_min, Z12_max
solve_Z12_range()

#Backtrack Z12 Function
def backtrack_Z12_solution(Z12_value):
    """
    Backtrack decision variable solution for a given Z12 value.
    """
    mdl = Model(name=f"PJK_Submodel_Z12={Z12_value}")

    # Define decision variables
    x = mdl.binary_var_dict(SR, name='x_')
    y = mdl.binary_var_dict(((r, f, m) for r in SR for f in SF for m in SM), name='y_')

    # Define constraints
    for s in SS:
        mdl.add_constraint(mdl.sum(alpha[r][s] * x[r] for r in SR) >= 1, ctname=f"min_coverage_{s}")                                        # Constraint 5
    mdl.add_constraint(mdl.sum(x[r] for r in SR) <= eta, ctname="max_routes")                                                               # Constraint 6
    for r in SR:
        for f in SF:
            for m in SM:
                mdl.add_constraint(m * y[r, f, m] <= lamda[r], ctname=f"cap_limit_{r}_{f}_{m}")                                             # Constraint 7
    for e in SE:
        mdl.add_constraint(mdl.sum(o[m] * beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM) >= omega[e], ctname=f"min_load_{e}") # Constraint 8
    for e in SE:
        mdl.add_constraint(mdl.sum(beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM) >= ae, ctname=f"min_flow_{e}")              # Constraint 9a
        mdl.add_constraint(mdl.sum(beta[r][e] * y[r, f, m] for r in SR for f in SF for m in SM) <= be, ctname=f"max_flow_{e}")              # Constraint 9b
    for r in SR:
        mdl.add_constraint(mdl.sum(y[r, f, m] for f in SF for m in SM) <= F * x[r], ctname=f"route_link_{r}")                               # Constraint 20
    #for f in range(1, F):
        #mdl.add_constraint(
            #mdl.sum(y[r, f + 1, m] for r in SR for m in SM) <= mdl.sum(y[r, f, m] for r in SR for m in SM),
            #ctname=f"sequential_flow_{f}")                                                                                                  # Constraint 21
    for f in SF:
        mdl.add_constraint(mdl.sum(y[r, f, m] for r in SR for m in SM) <= 1, ctname=f"max_flow_per_freq_{f}")                               # Constraint 22

    # Define objective as constraint
    obj = mdl.sum(c2[m] * g[r] * y[r, f, m] for r in SR for f in SF for m in SM)
    mdl.add_constraint(obj == Z12_value, ctname="fixed_Z12")
    # Solve the model
    mdl.parameters.lpmethod.set(method)
    solution = mdl.solve()

    # Check for feasibility
    if not solution:
        print("No feasible solution for Z12 =", Z12_value)
        return None

    # Print non-zero decision variables
    print("Solution found for Z12 =", Z12_value)
    #for var in mdl.iter_variables():
        #if var.solution_value != 0:
            #print(f"{var.name} = {var.solution_value}")
    if solution:
        variable_dict = {var.name: var.solution_value for var in mdl.iter_variables()}
        return variable_dict
    else:
        return None
    return variable_dict

#Z11 Function
def optimize_wsp(solution_Z12):
    """
    Optimize WSP using solution_Z12, computing Z11 and Z2.
    """
    if not solution_Z12:
        return None

    # Extract y values from the solution
    y = {}
    for key, value in solution_Z12.items():
        if key.startswith('y__'):
            index = key.split('__')[1].split('_')
            r, f, m = map(int, index)
            y[(r, f, m)] = value

    h = t / sum(y[(r, f, m)] for r in SR for f in SF for m in SM)
    findex = []
    for (r,f,m), value in y.items():
        if value >0:
            findex.append(f)
    findex.sort()
    fillf = list(set(SF)-set(findex))
    # Initialization
    S1 = [1, 8, 14, 24, 28, 35, 39]
    RTs = [0,300,0,0,0,0,0,0,300,0,0,0,0,0,300,0,0,0,0,0,0,0,0,0,300,0,0,0,300,0,0,0,0,0,0,300,0,0,0,210]
    Z11_star = float('inf')
    Z11_k = float('inf')
    cN = [0,0,0,0,0,0,0,0,0]
    z = 0
    # Optimization loop
    while Z11_k == float('inf'):
        gamma = []
        result = []
        DTs, ATs = {}, {}
        cN = [0,0,0,0,0,0,0,0,0]
        # Compute ATs and DTs
        for r in SR:
            for f in findex:
                for m in SM:
                    ATs[(1, map_origin.get((r, 1)), r, f, m)] = findex.index(f) * h * y[(r, f, m)]
                    ATs[(2, map_origin.get((r, 2)), r, f, m)] = (z + findex.index(f) * h) * y[(r, f, m)]
                    for s in SS:
                        if s != 1:
                            ATs[(1, s, r, f, m)] = ATs[(1, map_origin.get((r, 1)), r, f, m)] + (s - map_origin.get((r, 1))) * td[1] * y[(r, f, m)]
                        if s != S:
                            ATs[(2, s, r, f, m)] = ATs[(2, map_origin.get((r, 2)), r, f, m)] + (map_origin.get((r, 2)) - s) * td[2] * y[(r, f, m)]
        for r in SR:
            for f in findex:
                for m in SM:
                    for d in SD:
                        if d == 1:
                            for s in r_s1s2.get(r, []):
                                DTs[(d, s, r, f, m)] = ATs[(d, s, r, f, m)] + 15 * y[(r, f, m)]
                        if d == 2:
                            for s in r_s1s2.get(r, []):
                                DTs[(d, s, r, f, m)] = ATs[(d, s, r, f, m)] + 15 * y[(r, f, m)]


        DTs1 = {key: value for key, value in DTs.items() if value > 0  and value < 7201 and
               ((key[0] == 1 and key[1] in [map_origin.get((key[2],key[0]), [])]) or
                (key[0] == 2 and key[1] in [map_origin.get((key[2],key[0]), [])]))}

        ATs1 = {key: value for key, value in ATs.items() if value > 0 and value < 7201 and
               ((key[0] == 2 and key[1] in [map_destination.get((key[2],key[0]), [])]) or
                (key[0] == 1 and key[1] in [map_destination.get((key[2],key[0]), [])]))}

        # Iterate over each value in S1
        for s in S1:
            # Filter ATs1 and DTs1 based on the current value of s
            filtered_ATs1 = {key: value for key, value in ATs1.items() if key[1] == s}
            filtered_DTs1 = {key: value for key, value in DTs1.items() if key[1] == s}

            # Sort ATs1 by value (v) ascending
            sorted_ATs1 = sorted(filtered_ATs1.items(), key=lambda item: item[1])
            # Sort DTs1 by value (u) descending
            sorted_DTs1 = sorted(filtered_DTs1.items(), key=lambda item: item[1], reverse=True)

            # Process each element in sorted_ATs1
            for coorv, v in sorted_ATs1:
                # Initialize variables to track the minimum difference and corresponding U element
                min_diff = float('inf')
                selected_u_remove = None

                # Iterate over sorted_DTs1
                for cooru, u in sorted_DTs1:
                    if coorv[0] != cooru[0] and coorv[4] == cooru[4]:
                        diff = u - v - RTs[coorv[1]]
                        if diff >= 0 and diff < min_diff:  # Check for non-negative minimum difference
                            min_diff = diff
                            selected_u_remove = (cooru, u)
                            selected_v        = (coorv, v)
                # If a match is found, add it to the result and mark for removal
                if selected_u_remove:
                    (du,su,ru,fu,mu) = selected_u_remove[0]
                    (dv,sv,rv,fv,mv) = selected_v[0]
                    result.append((selected_u_remove,selected_v))
                    gamma.append(((du,ru,fu,rv,fv,mu)))
                    cN[mu] += 1
                    # Remove the selected U element from the sorted_DTs1 list
                    sorted_DTs1 = [item for item in sorted_DTs1 if item != selected_u_remove]
                    
    # Clear the delete list for the next iteration
        z += 10
        # Update objective
        Z11_star = sum(c1[m] * (2*sum(y[(r, f, m)] for r in SR for f in SF) - cN[m]) for m in SM)
        if Z11_star < Z11_k:
            Z11_k = Z11_star
        # Termination condition
        if not (2*sum(y[(r, f, 6)] for r in SR for f in SF) - cN[6] <= n[6] and
                2*sum(y[(r, f, 8)] for r in SR for f in SF) - cN[8] <= n[8]):
            Z11_k = float('inf')
        if z > 151:
            break
    # Output the resulting list
    print(f"Fixed Cost = {Z11_k}, offset = {z}, 6-trains = {sum(y[(r, f, 6)] for r in SR for f in SF)}, 8-trains = {sum(y[(r, f, 8)] for r in SR for f in SF)}, total circulation is = {sum(cN)} with {cN[6]} 6-trains and {cN[8]} 8-trains")
    return Z11_k,gamma,z,result

#Main Algorithm Function
def main_algorithm(theta):
    """
    Main algorithm to solve the combined optimization problem.
    """
    Z11 = 0
    Z12 = 0
    # Step 1: Solve for Z12_min and Z12_max
    Z12_min, Z12_max = solve_Z12_range()
    if Z12_min is None or Z12_max is None:
        print("Failed to find Z12 range.")
        return

    print(f"Z12_min: {Z12_min}, Z12_max: {Z12_max}")

    # Step 2: Iterate over Z12 values
    k = int((Z12_max- Z12_min)/theta)
    Z_star = float('inf')  # Initialize optimal value

    for step in range(0, k + 1):
        Z12_value = Z12_min + step * theta
        print(f"Solving for Z12 = {Z12_value}...")

        # Backtrack solution for Z12
        solution_Z12 = backtrack_Z12_solution(Z12_value)
        if not solution_Z12:
            print(f"Failed to solve PJK for Z12 = {Z12_value}.")
            continue

        # Solve WSP
        Z11_value, gamma, z, Z11_solution = optimize_wsp(solution_Z12)
        if Z11_value is None or Z11_solution is None:
            print(f"Failed to solve WSP or APM for Z12 = {Z12_value}.")
            continue
        # Solve APM submodels
        #Z2 = optimize_apm(solution_Z12, data_passengers)
        #if Z11 is None or Z2 is None:
            #print(f"Failed to solve WSP or APM for Z12 = {Z12_value}.")
            #continue
        # Compute overall objective Z
        Z_k = Z12_value + Z11_value
        print(f"{step}-th iteration, Z12 = {Z12_value}, Z11 = {Z11_value}, Total Z = {Z_k}")

        # Update Z_star
        if Z_k < Z_star:
            Z_star = Z_k
            Z11 = Z11_value
            Z12 = Z12_value

    print(f"Optimal Z*: {Z_star}")
    return Z_star,Z11,Z12
```
